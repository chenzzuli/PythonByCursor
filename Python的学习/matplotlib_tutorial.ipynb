```python
import numpy as np
import matplotlib.pyplot as plt
# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False    # 用来正常显示负号
# 创建示例数据
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制两条线
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--')

# 添加标题和标签
plt.title('正弦和余弦函数')
plt.xlabel('x 轴')
plt.ylabel('y 轴')

# 添加图例
plt.legend()

# 添加网格
plt.grid(True)

# 显示图形
plt.show()
```

```python
# 生成随机数据
np.random.seed(0)
x = np.random.normal(0, 1, 100)
y = 2 * x + np.random.normal(0, 0.5, 100)

# 创建散点图
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c='blue', alpha=0.5, label='数据点')

# 添加标题和标签
plt.title('散点图示例')
plt.xlabel('x 轴')
plt.ylabel('y 轴')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```

```python
# 准备数据
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')

# 在柱子上方添加数值标签
for i, v in enumerate(values):
    plt.text(i, v, str(v), ha='center', va='bottom')

# 添加标题和标签
plt.title('柱状图示例')
plt.xlabel('类别')
plt.ylabel('数值')

# 显示图形
plt.show()
```

```python
# 准备数据
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['lightblue', 'lightgreen', 'yellow', 'pink', 'orange']
explode = (0.1, 0, 0, 0, 0)  # 突出显示第一块

# 创建饼图
plt.figure(figsize=(10, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)

# 保持图形为圆形
plt.axis('equal')

# 添加标题
plt.title('饼图示例')

# 显示图形
plt.show()
```

```python
# 创建2x2的子图布局
plt.figure(figsize=(12, 10))

# 第一个子图：线形图
plt.subplot(2, 2, 1)
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x))
plt.title('线形图')

# 第二个子图：散点图
plt.subplot(2, 2, 2)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y)
plt.title('散点图')

# 第三个子图：柱状图
plt.subplot(2, 2, 3)
x = ['A', 'B', 'C', 'D']
y = [3, 7, 2, 4]
plt.bar(x, y)
plt.title('柱状图')

# 第四个子图：饼图
plt.subplot(2, 2, 4)
sizes = [30, 20, 25, 25]
plt.pie(sizes, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%')
plt.title('饼图')

# 调整子图间距
plt.tight_layout()

# 显示图形
plt.show()
```

```python
# 创建示例数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 创建图形
plt.figure(figsize=(12, 6))

# 设置线条样式
plt.plot(x, y, 
         color='blue',           # 线条颜色
         linestyle='--',         # 线条样式
         linewidth=2,            # 线条宽度
         marker='o',             # 标记样式
         markersize=6,           # 标记大小
         markerfacecolor='red',  # 标记填充颜色
         label='sin(x)')         # 图例标签

# 设置标题和轴标签
plt.title('样式设置示例', fontsize=16, pad=20)
plt.xlabel('X 轴', fontsize=12)
plt.ylabel('Y 轴', fontsize=12)

# 设置坐标轴范围
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# 添加网格
plt.grid(True, linestyle=':', alpha=0.6)

# 添加图例
plt.legend(fontsize=10)

# 显示图形
plt.show()
```

```python
# 创建一个简单的图表
plt.figure(figsize=(8, 6))
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.title('保存图表示例')

# 保存为PNG格式（也支持jpg、pdf等格式）
plt.savefig('matplotlib_example.png', 
            dpi=300,             # 分辨率
            bbox_inches='tight', # 去除多余的白边
            pad_inches=0.1)      # 设置边距

plt.show()
```
